```python

import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import cartopy.feature as cfeature
```

```python
data = xr.open_dataset('Africa_temp.nc')
data
```

```python
tmp= data['tmp']
tmp
```

```python
yearly = tmp.groupby('time.year').mean('time')
yearly
```

```python
monthly = tmp.groupby('time.month').mean('time')
monthly
```

```python
yearly.plot(col='year', col_wrap= 10)
```

```python
monthly.plot(col='month', col_wrap=4,figsize=(10,10) ,vmin= 0)

```
